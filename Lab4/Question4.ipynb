```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Define the function y(x)
def f(x):
    return 2*x*np.exp(x**2)

# Define the exact area under the curve
def exact_area():
    return np.exp(9) - np.exp(1)

# Define the trapezoidal rule for integration
def trapezoidal_rule(f, a, b, M):
    h = (b - a) / M
    integral = 0.5 * (f(a) + f(b))
    for i in range(1, M):
        integral += f(a + i * h)
    integral *= h
    return integral

# Define the interval [a, b]
a, b = 1, 3

# List of number of intervals M
M_values = np.arange(2, 101, 1)

# Compute the area under the curve for each M
areas = [trapezoidal_rule(f, a, b, M) for M in M_values]

# Exact area under the curve
exact = exact_area()
```

```python
# Plot the results

# plt.figure(figsize=(10, 6))
plt.plot(M_values, areas, label='Area (Trapezoidal Rule)')
plt.hlines(exact, M_values[0], M_values[-1], colors='r', linestyles='dashed', label='Exact Area')
plt.xlabel('Number of Intervals (M)')
plt.ylabel('Area under the curve')
plt.title('Visualisation of Area under the curve as a function of M')
plt.legend()
plt.grid(True)
plt.show()
```
