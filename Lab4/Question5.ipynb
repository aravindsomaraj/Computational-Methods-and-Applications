```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Define the function y(x)
def y(x):
    return 2 * x * np.exp(x**2)

# Define the exact area under the curve
def exact_area(u):
    return np.exp(u**2) - np.exp(0)
```

```python
# Define the upper limits of integration (u values)
u_values = np.linspace(0.1, 2, 100)

# Compute the area under the curve for each u using various integration methods
area_trapz = [integrate.trapz(y(np.linspace(0, u, 100)), np.linspace(0, u, 100)) for u in u_values]
area_simps = [integrate.simps(y(np.linspace(0, u, 100)), np.linspace(0, u, 100)) for u in u_values]
area_quad = [integrate.quad(y, 0, u)[0] for u in u_values]
fixed_quad_areas = [integrate.fixed_quad(y, 0, u)[0] for u in u_values]
quadrature_areas=[integrate.quadrature(y, 0, u)[0] for u in u_values]
```

```python
#Now plot them as usual
plt.figure(figsize=(10, 6))
plt.plot(u_values,area_quad,label='quad',c='r', zorder = 6)
plt.plot(u_values,fixed_quad_areas,label='fixed_quad',c='g', linewidth=5, zorder = 4)
plt.plot(u_values,area_simps,label='simps',c='yellow',linewidth=8, zorder=2)
plt.plot(u_values,quadrature_areas,label='quadrature',c='orange', linewidth=11, zorder=0)
# plt.plot(u_values,exact_area(u_values),label='Actual area under the curve',c='b')  #Indicating the actual area under the curve
plt.xlabel('u')
plt.ylabel('Area under the curve')
plt.title('Area under 2xe^(x^2) in the interval [0,u]')
plt.legend()
plt.grid()
plt.show()
```
