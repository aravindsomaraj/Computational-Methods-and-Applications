```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the function sin(x^2)
def f(x):
    return np.sin(x**2)

# Defining the actual derivative of f(x)
def f_dash(x):
    return 2*x*np.cos(x**2)

# Defining the <forward,backward,centered> finite difference approximation of f(x)
def delta_plus(x, h):
    return (f(x + h) - f(x)) / h
def delta_minus(x, h):
    return (f(x) - f(x - h)) / h
def delta_centered(x, h):
    return (f(x + h) - f(x - h)) / (2*h)

```

```python
x_values = np.linspace(0, 1, 100)
y_values = f_dash(x_values)
y_delta_plus = delta_plus(x_values, 0.01)
y_delta_minus = delta_minus(x_values, 0.01)
y_delta_centered = delta_centered(x_values, 0.01)

y_delta_plus_absError = abs(y_values - y_delta_plus)
y_delta_minus_absError = abs(y_values - y_delta_minus)
y_delta_centered_absError = abs(y_values - y_delta_centered)

plt.plot(x_values, y_delta_plus_absError, label="Absolute error in forward difference")
plt.plot(x_values, y_delta_centered_absError, label="Absolute error in central difference")
plt.plot(x_values, y_delta_minus_absError, label="Absolute error in backward difference")
plt.legend()
plt.title("Absolute errors in forward, central, backward difference methods for " +r"$sin(x^2)$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
