## Monte Carlo Estimation of $\pi$

```python
import matplotlib.pyplot as plt
import random
import math as mt
```

### Estimation of $\pi$
- Consider a unit square centered at the origin of a 2-D plane.
- Inscribe a circle within this square.
- Generate points of type $(x, y)$, where $x$ and $y$ are uniformly and independently sampled from the interval $[âˆ’0.5, 0.5]$, and place them on the 2-D plane.
- Then, for a large number points, we have   
$$\frac{no.\ points\ within\ the\ circle}{no.\ points\ within\ the\ square} \stackrel{(a)}{=} \frac{area\ of\ circle}{area\ of\ square} = \frac{\pi}{4}$$ 
- Or, equivalently,  
    $$\pi = \frac{4 \times no.\ points\ within\ the\ circle}{no.\ points\ within\ the\ square}$$
    

```python
# Estimation of pi using Monte Carlo method
def estimatePi(n):
    if not isinstance(n, int) or n <= 0:
        raise ValueError("Number of points should be a positive integer.")

    points_inside_circle = 0
    pi_estimates = []

    for i in range(1, n + 1):
        x, y = random.uniform(0, 1), random.uniform(0, 1)
        distance_squared = x**2 + y**2

        if distance_squared <= 1:
            points_inside_circle += 1

        pi_estimate = 4 * points_inside_circle / i
        pi_estimates.append(pi_estimate)

    return pi_estimates
```

```python
total_points = 10000

# Estimate Ï€ using Monte Carlo method
pi_estimates = estimatePi(total_points)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, total_points + 1), pi_estimates, label='Monte Carlo Method')
plt.axhline(y=mt.pi, color='r', label='Value of math.pi')
plt.xlabel('Number of Points Generated')
plt.ylabel('Estimated Value of Ï€')
plt.title('Estimating Ï€ using Monte Carlo Method')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```
