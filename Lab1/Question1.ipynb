```python
import numpy as np 
import matplotlib.pyplot as plt
import math as mt
from math import factorial, sqrt, pi, exp
```

```python
lim = 50     # Setting limit for the factorial function
dp = np.zeros((lim-1,))
x = np.arange(1, lim)

# Since factorial values are very large, we use log10 to store them, using dynamic programming to further improve speed
def Fact(i):
    dp[i-1] = dp[i-2] + np.log10(i)
    return dp[-1]
```

```python

for i in range(1,lim):
    Fact(i)
# print(dp)
```

```python
stirling = np.zeros((lim-1,))
def Stirlings(i):
    # Using Stirling's approximation to calculate the factorial, after taking log
    return (0.5*mt.log(2*pi*i,10)) + (i*mt.log((i/mt.e),10))
for i in range(1,lim):
    stirling[i-1] = Stirlings(i)
```

```python
plt.plot(dp, label = 'Actual factorial (log)', color = 'yellow', linewidth=2.5)
plt.plot(stirling, label = 'Stirling approximation (log)', color = 'black', linestyle = '--')

plt.xlabel('n')
plt.ylabel('log(n!)')
plt.title('Stirling approximation vs Actual factorial')

plt.legend()
plt.show()
```
